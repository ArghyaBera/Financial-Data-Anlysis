```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Financial Analytics data.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['S.No.','Unnamed: 4'])
```

```python
df.head()
```

```python
top=df.sort_values(by='Mar Cap - Crore',ascending=False)
topp=top[['Name', 'Mar Cap - Crore']]

plt.figure(figsize=(8, 4))
plt.plot(topp['Mar Cap - Crore'])
plt.title('Mar Cap - Crore')
plt.ylabel('Mar Cap - Crore')
plt.grid(True)

plt.show()
```

Heat Map

```python
correlation_matrix = df[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Market Cap vs Quarterly Sales')
plt.show()
```

Top 10 Companies based on Market Capitilazation

```python
top_10_mar=df.sort_values(by='Mar Cap - Crore',ascending=False).head(10)
top10=top_10_mar[['Name', 'Mar Cap - Crore']]
top10
```

```python
company_names = top_10_mar['Name']
mar_cap_values = top_10_mar['Mar Cap - Crore']
plt.figure(figsize=(12, 6))

plt.pie(mar_cap_values, labels=company_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Companies Market Capitalization Distribution')
```

```python
x=top10['Name']
y=top10['Mar Cap - Crore']
plt.plot(x,y,marker='o')
plt.xticks(rotation=80)
```

```python
topqdf = df.sort_values(by='Sales Qtr - Crore', ascending=False)
topf = topqdf[['Name', 'Sales Qtr - Crore']]

plt.figure(figsize=(8, 4))
plt.plot(topf['Name'], topf['Sales Qtr - Crore'])
plt.title('Sales Qtr - Crore')
plt.ylabel('Sales Qtr - Crore')
plt.xticks([])
plt.show()
```
